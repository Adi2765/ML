```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#MATPLOTLIB BASICS
x=np.linspace(1,10,10)
y=x**2
print(x)
print(y)
plt.plot(x,y)      # creates a graph of x vs y
plt.title("Example")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
```

```python
#SUBPLOTS -->HAVE TO GIVE 3 ARGUMENTS INTO IT ...THE FIRST TWO REPRESENTS THE GRID(ROWS X COLUMNS) IN WHICH THE WHOLE SPACE IS DIVIDED AND THE 3RD ARG REPERESENTS THE PART OF GRID TO BE ACCESSED
plt.subplot(2,2,1)  #-->access first subpart/subplot of 2x2 plot
plt.plot(x,y)     #--->plot in the space choosen above
plt.subplot(2,2,3)
plt.plot(y,x) 
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(y,x**2)
```

```python
#OBJECT CREATION:INVOLVES CREATING AN OBJECT HAVING THE WHOLE CANVAS AND PLOTTING GRAPHS ACCORDINGLY. 
fig=plt.figure()
object1=fig.add_axes([0.1,0.1,.8,.8])   #all 4 values range from 0 to 1..1st:distance from origin in x-axis...2nd:distance from origin in y-axis...3rd:width,4th:height
plt.plot(x,y)    #plotting will be done on the canvas provided above
object2=fig.add_axes([0.3,0.4,0.2,0.2])  #overlapping the first plot on the canvas
plt.plot(y,x)
#figure size and dpi(dots per inch or increasing pixel) can also be modified using fig=plt.figure(figsize=(6,8),dpi=100(defaut) ) --->means 600x800
```

```python
#TYPES OF PLOTS
#1->SCATTER PLOTS 
plt.scatter(x,y)  # scatter in the form of points in the graph
```

```python
#2->histogram 
plt.hist(x,y)
```

```python
#3->BOXPLOTS 
plt.boxplot(x)
```

```python
#saving plots
fig=plt.figure(figsize=(6,8),dpi=200)
fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y)
plt.savefig("fig1.png")
```

```python
#designing of plots 
plt.plot(x,y,color='red',linewidth=1,linestyle='+')
```
