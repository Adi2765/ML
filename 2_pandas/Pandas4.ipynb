```python
#GROUP BY 
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
df
```

```python
#df.groupby creates an object so it doesn't show on printing and the objects contain the original info divided into categories
cat=df.groupby('Category')
for i,j in cat:
    print(i)
    print(j)
```

```python
#group the data by category and calculate the sum for sales in each.
cat=df.groupby('Category')['Sales'].sum()
print(cat)

#group by store and calculate the sum of sales.
cat1=df.groupby('Store')['Sales'].sum()
print(cat1)

#group by multiple columns(category and store)
cat2=df.groupby(['Category','Store'])['Sales'].sum()
print(cat2)
```

```python
#AGGREGATION
print(df['Sales'].mean())
print(df['Sales'].median())
print(df['Sales'].mode()[0])
print(df['Sales'].min())
print(df['Sales'].max())
print(df['Sales'].count())
print(df['Sales'].std())
print()

#MULTIPLE FUNC IN A SINGLE LINE
print(df['Sales'].agg(['mean','median','max']))
```

```python
#PIVOT TABLE
data1 = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data1)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
print(df)
```

```python
#creates a modified table with separate desired values,columns and indexes.
print(pd.pivot_table(df,values='Sales',index='Region',columns='Product')) #by default the values fill up with mean which can be modified
print()
print(pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median'))
```

```python
#crosstab:gives out the count
pd.crosstab(df['Region'],df['Product'])
```
