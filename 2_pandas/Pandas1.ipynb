```python
Series:1D labelled array capable of holding any datatype..axis labels are collectively called the index.
```

```python
# Different ways of creating a series in python.
import numpy as np
import pandas as pd
labels=['a','b','c']
list=[10,20,30]
d={"adi":20,"ayush":18,"aman":19}
arr=np.array([1,2,3,4,5])
s=pd.Series(list,index=labels)
s1=pd.Series(d)
s2=pd.Series(arr)
print(s)
print(s1)
print(s2)
```

```python
# DataFrame:2D labelled data structure with rows and columns capable of storing different data types in each column
```

```python
#Dataframe representation
import numpy as np
import pandas as pd
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
columns=['Name','Age','City','Salary']
df1=pd.DataFrame(data_list,columns=columns)
print(df1)
print()
#selecting an entire single column
print(df1['Name'])
print()
#selecting two columns at once 
print(df1[['Name','City']])
print()
#creating a new column
df1['designation']=['doctor','eng','doctor','eng'] #-->fill the list with the exact no of values required.......
print(df1)
print()
#removing a column
print(df1.drop('designation',axis=1))# ---->axis works oppositely in comparison to numpy arrays....axis=0 means row-wise and 1 means column wise
print()
print(df1.drop('designation',axis=1,inplace=True)) #---->inplace keyword permanently removes the column from the memory and not just for show case purpose.
print()
#removing rows involves only changing the axis from 1 to 0
print(df1.drop(0,axis=0))
print()
#selecting rows 
print(df1.loc[0])
print()
print(df1.loc[[1,2]])   # iloc is used when the indexes aren't labelled as 0,1,2,....
print()
#selecting a random grid of values/subsets of rows and columns-->start with rows using loc and then mention the columns need in the specific rows
print(df1.loc[[0,1]][['Name','Salary']]) #-->selects the rows 0 and 1 alongwith the columns name and salary
print(df1.loc[[1,2]][['City','Salary']])

#loc is used in label based indexing whereas iloc is used in numerical based indexing
```

```python
# CONDITIONAL SELECTION
import pandas as pd
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
columns=['Name','Age','City','Salary']
df=pd.DataFrame(data_list,columns=columns)
#SELECT ONLY THOSE PEOPLE WHO'S AGE IS ABOVE 30
print(df[df['Age']>30]) 
print()
#AGE ABOVE 30 AND CITY IS 'PARIS'
print(df[(df['Age']>30) & (df['City']=='Paris')])
```
